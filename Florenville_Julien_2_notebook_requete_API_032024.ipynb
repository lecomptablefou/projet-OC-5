```python
from stackapi import StackAPI
import pandas as pd
from datetime import datetime

# Convertir les dates en timestamps Unix
from_date = int(datetime(2022, 1, 1).timestamp())
to_date = int(datetime(2023, 5, 31, 23, 59, 59).timestamp())

# Créer une instance StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=from_date, todate=to_date, tagged='python', min=50, sort='votes', order='desc', pagesize=50)

# Extraire les données nécessaires
questions_data = []
for item in questions['items']:
    question_info = {
        'Date': pd.to_datetime(item['creation_date'], unit='s'),
        'Titre': item['title'],
        'Tags': item['tags'],
        'Score': item['score']
    }
    questions_data.append(question_info)

# Créer un DataFrame
df_questions = pd.DataFrame(questions_data)

# Afficher le DataFrame
df_questions
```

```python
# Enregistrer le DataFrame en CSV
df_questions.to_csv('questions_python.csv', index=False)
```
